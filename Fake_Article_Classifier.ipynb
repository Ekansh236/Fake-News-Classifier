```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
```

```python
df = pd.read_csv("fake_or_real_news.csv")
```

```python
df
```

```python
def map_label(label):
    if label == 'REAL':
        return 0
    elif label == 'FAKE':
        return 1

df['label'] = df['label'].apply(map_label)

# Print the updated DataFrame
print(df['label'])
```

```python
df
```

```python
x = df['text']
y = df['label']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
vectorizer = TfidfVectorizer(stop_words = "english", max_df = 0.7)
x_train_vectorized = vectorizer.fit_transform(x_train)
x_test_vectorized = vectorizer.transform(x_test)
```

```python
model = LinearSVC()
model.fit(x_train_vectorized, y_train)
```

```python
model.score(x_test_vectorized, y_test)
```

```python
def classify_text(user_input):
  user_input = vectorizer.transform([user_input])
  prediction = model.predict(user_input)
  return prediction[0]

user_input = input("Enter text you would like to classify: ")
result = classify_text(user_input)
if result == 1:
  print("This article is Fake")
else:
  print("This article is Real")
```
